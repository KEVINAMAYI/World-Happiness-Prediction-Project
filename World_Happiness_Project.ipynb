*HERE IS THE CASE*

1. POSTIVIVE CORRELATION

IF TWO VALUES HAVE

a.) A CORRELATION VALUE OF BETWEEN (0.5 to 1) => THEY ARE STRONGLY POSITIVELY CORRELATED 
(That is you increase the value of one the other increases relatively with the same degree to that percentage increase, and vice versa,
you decrease one, the other one decease relatively to that percentage) e.g eating healthy and being health

b.) A CORRELATION VALUE OF BETWEEN (0.5 to 0) => THEY ARE LOOSELY POSITIVELY  CORRELATED
(That is you increase the value of one the other increases with a very very very  small percent or no increase at all,
 and vice versa,you decrease one, the other one decease with a very very very  small percent) e.g reading code and coding skills


2. NEGATIVE CORRELATION

IF TWO VALUES HAVE

3.) A CORRELATION VALUE OF BETWEEN (-0.5 to -1) => THEY ARE STRONGLY NEGATIVELY CORRELATED 
(That is you increase the value of one the other decreases relatively to that percentage increase, and vice versa,
you decrease one, the other one increases relatively to that percentage) e.g Love and Hatred

4.) A CORRELATION VALUE OF BETWEEN (-0.5 to 0) => THEY ARE LOOSELY NEGATIVELY  CORRELATED
(That is you increase the value of one the other decreases with a very very very  small percent or no decrease at all, 
 and vice versa, you decrease one, the other one increases with a very very very  small percent) e.g writing and travelling

# Model Building

```python
# Choose an algorithm
# Get X and Y variables

#Store Dependent variable
Y = world_happiness_file.loc[:,'Score'].values

#Store Independent variable
X = world_happiness_file.loc[:,('GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices')].values

#Store Independent variable
X2 = world_happiness_file.loc[:,('GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices')]
```

```python
#Split data into training and test data sets => First Method
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.25,random_state=0) 
```

```python
# Split data into training and test data sets => Second Method

# shuffled_data = world_happiness_file.sample(frac=1)
# Xtrain = shuffled_data.loc[:100,('GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices')].values

# Xtest = shuffled_data.loc[100:,('GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices')].values

# Ytrain = shuffled_data.loc[:100,('Score')].values
# Ytest  = shuffled_data.loc[100:,('Score')].values


```

```python
#importing linear model algorithm
from sklearn import linear_model
```

```python
#creating and training the model
regression = linear_model.LinearRegression()

regression.fit(X_train,y_train)
```

```python
#testing the model

y_prediction = regression.predict(X_test)
y_prediction
```

```python
prediction_result = pd.DataFrame({
    'Actual Data':y_test,
    'Predicted Data':y_prediction
})

prediction_result['Difference'] = y_test - y_prediction
prediction_result.head()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

print('Mean Absolute Error',mean_absolute_error(y_test,y_prediction))
print('Mean Squared Error',mean_squared_error(y_test,y_prediction))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(y_test,y_prediction)))

#THE SCORE OF THE MODEL
print('Variance %0.2f' %(regression.score(X_test,y_test)))
```

```python
#Sketch the Actual values against the predicted values
sns.regplot(x='Actual Data',y='Predicted Data', data=prediction_result)
```

```python
# In linear Regression the Model is Just a Function
# Y = MX + C => M is called the Coefficient and C the intercept

# Y = M1*X1 + M2*X2 + M3*X3 + M4*X4 ... + C 

# (IN OUR CASE)
# Hapiness_Score = M1 * GDP per Capita + M2 * Social support + M3 * Healthy life expectancy + m4 * Freedom to make life choices

regression.intercept_
```
